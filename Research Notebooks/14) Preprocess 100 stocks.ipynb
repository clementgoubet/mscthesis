```python
import dask.dataframe as dd
import pandas as pd
import numpy as np
```

```python
Xh5 = dd.read_hdf('/Users/clement/DataArrow/X_normalized.h5', "us")
```

```python
cols = list(Xh5.columns[:-2])
print(cols)
```

```python
print(Xh5)
Xh5[cols] = (Xh5[cols]-Xh5[cols].mean())/Xh5[cols].std()
Xh5['y'] = Xh5['Ret']/abs(Xh5['Ret'])
Xh5 = Xh5.dropna()
```

```python
X = Xh5.compute()
```

```python
print(X)
import pickle
with open('/Users/clement/DataArrow/X_normalized.pickle', 'wb') as handle:
    pickle.dump(X, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
dd.to_hdf(Xh5,'/Users/clement/DataArrow/X_normalized.h5', "us")
```
